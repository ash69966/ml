```python
df = df.drop(['index', 'encounter_id', 'race', 'medical_specialty', 'weight'], axis=1)
```

```python
Age = []
for i in df['age']:
  if i == '[80-90)':
    Age.append(85)
  elif i == '[60-70)':
    Age.append(65)
  elif i == '[50-60)':
    Age.append(55)
  elif i == '[70-80)':
    Age.append(75)
  elif i == '[40-50)':
    Age.append(45)
  elif i == '[20-30)':
    Age.append(25)
  elif i == '[30-40)':
    Age.append(35)
  elif i == '[90-100)':
    Age.append(95)
  elif i == '[0-10)':
    Age.append(5)
  else:
    Age.append(15)
```

```python
df['age'] = Age
```

```python
#Creating a Model 
x = df.drop('readmitted', axis=1)
y = df['readmitted']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
predictions = clf.predict(x_test)
```

```python
accuracy = confusion_matrix(y_test, predictions)
accuracy
```
